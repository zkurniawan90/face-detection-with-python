## Import Modules

Berikut adalah beberapa Python modules yang perlu kita import untuk dapat mengikuti tutorial ini.

````python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
````

## Global Identifiers

Dalam tutorial ini terdapat beberapa file foto yang dapat digunakan untuk mempelajari deteksi wajah.

````python
IMAGE_PATH = './dataset/classroom/03.jpg'
# IMAGE_PATH = './dataset/babies/03.jpg'
````

## Load Image Dataset

Membuat objek image OpenCV

````python
bgr_img = cv2.imread(IMAGE_PATH)
````

````python
bgr_img.shape
````

````python
bgr_img
````

````python
plt.figure(figsize=(5, 5))
plt.imshow(bgr_img)
````

## RGB Format

````python
rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
````

````python
plt.figure(figsize=(5, 5))
plt.imshow(rgb_img)
````

## Gray Scale Image

````python
gray_img = cv2.cvtColor(src=rgb_img, code=cv2.COLOR_RGB2GRAY)
````

````python
plt.figure(figsize=(5, 5))
plt.imshow(gray_img)
````

````python
plt.figure(figsize=(5, 5))
plt.imshow(cv2.cvtColor(src=gray_img, code=cv2.COLOR_GRAY2RGB))
````
